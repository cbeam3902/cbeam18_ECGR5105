## Problem 1

```python
# Setup
import torch
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Get dataset
df = pd.read_csv("../datasets/D3.csv")
df
```

```python
# Linear Regression and Gradient_descent

def lineReg(X, Y, theta):
    pred = X.dot(theta)
    errors = np.square(np.subtract(pred,Y))
    return 1 / (2 * len(Y)) * np.sum(errors)

def gradDes(X, Y, theta, alpha, iterations):
    costArray = np.zeros(iterations)

    for i in range(iterations):
        pred = X.dot(theta)
        errors = np.subtract(pred, Y)
        delta = (alpha / len(Y)) * X.transpose().dot(errors)
        theta = theta - delta
        costArray[i] = lineReg(X, Y, theta)
    return theta, costArray
```

```python
# Collect the Y Values
Y = df.values[:,3]
tempOnes = np.ones((len(Y),1))
iterations = 100
```

### X1 and Y line fit

```python
# Setup the X Values to match b + mx format [1, x]
X1_val = df.values[:,0]
X1_val = X1_val.reshape(len(Y), 1)
X1 = np.hstack((tempOnes, X1_val))
```

```python
# Perform regression

learnSize = 20
learnRange = np.linspace(0.01, 0.1, learnSize, endpoint=True)
errorOut = []
thetaArray = []
lossArray = []
for learnRate in learnRange:
    theta_x1 = np.zeros(2)
    theta_x1, loss = gradDes(X1, Y, theta_x1, learnRate, iterations)
    lossArray.append(loss)
    thetaArray.append(theta_x1)
    errorOut.append(lineReg(X1, Y, theta_x1))

idx = np.argmin(errorOut)
print("Learn rate with lowest error is {:.1f}".format(learnRange[idx]))
print("Variables Found:\n\ttheta0 = {:.5f}\n\ttheta1 = {:.5f}".format(thetaArray[idx][0], thetaArray[idx][1]))
plt.scatter(X1[:,1], Y, color='red', marker= '+', label = 'Training Data')
plt.plot(X1[:,1], X1.dot(thetaArray[idx]), color='green', label='Linear Regression')

plt.title("X1 Line Fit")
plt.ylabel("Values")
plt.xlabel("X1")
plt.grid()
plt.legend()
plt.show()

plt.plot(range(iterations),lossArray[idx])
plt.title("X1 Loss")
plt.ylabel("Loss")
plt.xlabel("Iteration")
plt.grid()
plt.show()
```
