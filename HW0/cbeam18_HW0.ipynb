```python
# Setup
import torch
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Get dataset
df = pd.read_csv("../datasets/D3.csv")
df
```

```python
# Linear regression and it's derivative (1 variable)

def lineReg(x, y, w, b):
    error = (x * w + b - y) ** 2
    error = np.sum(error)
    return error/(2 * len(x))

def lineReg_der(x, y, w, b):
    error_w = x * (x * w + b - y)
    error_w = np.sum(error_w)
    
    error_b = (x * w + b - y)
    error_b = np.sum(error_b)

    return error_w/len(x), error_b/len(x)
```

```python
# learnRate = 0.1
learnSize = 20
learnRange = np.linspace(0.01, 0.1, learnSize, endpoint=True)
errorOut = []
xArray = []
bArray = []
lossArray = []
for learnRate in learnRange:
    x1 = 0
    b = 0
    loss = []
    for iii in range(100):
        error_x, error_b = lineReg_der(df.X1.values, df.Y.values, x1, b)
        x1 = x1 - learnRate * error_x
        b = b - learnRate * error_b
        loss.append(lineReg(df.X1.values, df.Y.values, x1, b))
    lossArray.append(loss)
    errorOut.append(lineReg(df.X1.values, df.Y.values, x1, b))
    xArray.append(x1)
    bArray.append(b)

idx = np.argmin(errorOut)
print("Learn rate with lowest error is {:.1f}".format(learnRange[idx]))
print("Variables Found:\n\tX1 = {:.5f}\n\tb = {:.5f}".format(xArray[idx], bArray[idx]))
plt.scatter(df.X1.values, df.Y.values)
plt.plot(df.X1.values, xArray[idx] * df.X1.values + bArray[idx])
plt.title("X1 Error")
plt.ylabel("Error")
plt.xlabel("X1")
plt.show()

plt.plot(range(100),lossArray[idx])
plt.title("X1 Loss")
plt.ylabel("Loss")
plt.xlabel("Iteration")
plt.show()
```

```python
# Linear Regression (3 variables)

def lineReg3(x1, x2, x3, y, w1, w2, w3, b):
    error = (x1 * w1 + x2 * w2 + x3 * w3 + b - y) ** 2
    error = np.sum(error)
    return error/(2 * len(x1))

def lineReg3_der(x1, x2, x3, y, w1, w2, w3, b):
    error_w1 = x1 * (x1 * w1 + x2 * w2 + x3 * w3 + b - y)
    error_w1 = np.sum(error_w1)

    error_w2 = x2 * (x1 * w1 + x2 * w2 + x3 * w3 + b - y)
    error_w2 = np.sum(error_w2)

    error_w3 = x3 * (x1 * w1 + x2 * w2 + x3 * w3 + b - y)
    error_w3 = np.sum(error_w3)
    
    error_b = (x1 * w1 + x2 * w2 + x3 * w3 + b - y)
    error_b = np.sum(error_b)

    return error_w1/len(x1), error_w2/len(x1), error_w3/len(x1), error_b/len(x1)
```

```python
learnSize = 20
learnRange = np.linspace(0.01, 0.1, learnSize, endpoint=True)
errorOut = []
x1Array = []
x2Array = []
x3Array = []
bArray = []
lossArray = []
for learnRate in learnRange:
    x1 = 0
    x2 = 0
    x3 = 0
    b = 0
    loss = []
    for iii in range(100):
        error_x1, error_x2, error_x3, error_b = lineReg3_der(df.X1.values, df.X2.values, df.X3.values, df.Y.values, x1, x2, x3, b)
        x1 = x1 - learnRate * error_x1
        x2 = x2 - learnRate * error_x2
        x3 = x3 - learnRate * error_x3
        b = b - learnRate * error_b
        loss.append(lineReg3(df.X1.values, df.X2.values, df.X3.values, df.Y.values, x1, x2, x3, b))
    lossArray.append(loss)
    errorOut.append(lineReg3(df.X1.values, df.X2.values, df.X3.values, df.Y.values, x1, x2, x3, b))
    x1Array.append(x1)
    x2Array.append(x2)
    x3Array.append(x3)
    bArray.append(b)

idx = np.argmin(errorOut)
print("Learn rate with lowest error is {:.1f}".format(learnRange[idx]))
print("Variables Found:\n\tX1 = {:.5f}\n\tX2 = {:.5f}\n\tX3 = {:.5f}\n\tb = {:.5f}".format(x1Array[idx], x2Array[idx], x3Array[idx], bArray[idx]))

plt.plot(range(100),lossArray[idx])
plt.title("Total Loss")
plt.ylabel("Loss")
plt.xlabel("Iteration")

print("Y values:\n\t(1, 1, 1) = {:.5f}\n\t(2, 0, 4) = {:.5f}\n\t(3, 2, 1) = {:.5f}".format(x1 + x2 + x3 + b, 2 * x1 + 4 * x3 + b, 3 * x1 + 2 * x2 + x3 + b))
```
